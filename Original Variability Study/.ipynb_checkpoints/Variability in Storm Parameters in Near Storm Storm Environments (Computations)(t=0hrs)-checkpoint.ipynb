```python
#Need to Specify Bounds For ~25km x 25km Box (Pick It To Be in the Middle of Domain)

midpoint = round(len(x_data)/2)-1 #Roughly the Middle of the Domain (Not Exact Because Even Number of Points)
slope = (x_data[2]-x_data[1]).values #Distance Between Each Point
num_points = round(25/2/slope)
upper_bound = midpoint + num_points + 1
lower_bound = midpoint - num_points 
horiz_length = upper_bound-lower_bound
# Load necessary data only once
storm_data = xr.open_dataset(path2 + storm16_files[0])
```

```python
#CAPE-CIN (ALL AT ONCE)

#CAPE-CIN

#Import Data
QV = storm_data['qv'].loc[:,:,lower_bound:upper_bound,lower_bound:upper_bound][0] #Assume SH~Water Vapor Mass Fraction (Ask Paul if Correct)
P = storm_data['prs'].loc[:,:,lower_bound:upper_bound,lower_bound:upper_bound][0]/100 #pressure at gridpoint
Th = storm_data['th'].loc[:,:,lower_bound:upper_bound,lower_bound:upper_bound][0] #potential temp at gridpoint

#For the WRF CAPE Function We Need: Total Pressure (hPa), Temperature (degK), Water Vapor Mixing Ratio (kg/kg), *Geopotential Height* (m), Terrain Height (m), Surface Pressure (hPa)
T = mpcalc.temperature_from_potential_temperature(P*units.hPa,Th*units.kelvin) #Temperature from Potential Temperature
z = np.asarray(z_data*1000).reshape(121,1,1) #"Geopotential Height"
Height = np.tile(z, (1, horiz_length, horiz_length)) #"Geopotential Height"
Terrain = np.full((horiz_length,horiz_length),0) #Terrain (Assuming Flat Surface)
P_sfc = P[0].values #Surface Pressure

#WRF CAPE Computation
results = wrf.cape_3d(P,T,QV,Height,Terrain, P_sfc, ter_follow = True,meta= True)
```

```python
#Plot CAPE Field

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

plt.pcolormesh(x_data[lower_bound:upper_bound], y_data[lower_bound:upper_bound], results[0][0])
plt.title('Variability of CAPE in 25 km x 25km Domain', fontsize = 16)
plt.xlabel('Grid Spacing (km)')
plt.ylabel('Grid Spacing (km)')
clb = plt.colorbar()
clb.set_label('J/kg', fontsize = 13)
```

```python
#SRH1km and SRH3km

#Import Data
u = storm_data['uinterp'].loc[:,:,lower_bound:upper_bound,lower_bound:upper_bound][0]
v = storm_data['vinterp'].loc[:,:,lower_bound:upper_bound,lower_bound:upper_bound][0]

#For the WRF SRH Function We Need: U (m/s), V (m/s), Height (m), Terrain (m), Top (m)
Height = Height
Terrain = np.full((horiz_length,horiz_length),0)

SRH1km = wrf.srhel(u,v,Height, Terrain, top = 1000)
SRH3km = wrf.srhel(u,v,Height, Terrain, top = 3000)

```

```python
#METPY CALCULATIONS SCHEME: WAY TOO SLOW!

def computations(x,y):
    #CAPE-CIN
    SH = xr.open_dataset(path2+storm16_files[0])['qv'].loc[:,:,y,x] #Assume SH~Water Vapor Mass Fraction (Ask Paul if Correct)
    P = xr.open_dataset(path2+storm16_files[0])['prs'].loc[:,:,y,x] #pressure at gridpoint
    Th = xr.open_dataset(path2+storm16_files[0])['th'].loc[:,:,y,x] #potential temp at gridpoint
    T = mpcalc.temperature_from_potential_temperature(P*units.Pa,Th*units.kelvin)
    Td = (mpcalc.dewpoint_from_specific_humidity(P*units.Pa, T, SH*units('kg/kg')).values +273.15)*units.kelvin #dewpoint via metpy
    Parc_prof = mpcalc.parcel_profile(P[0].values*units.Pa,T[0][0].values*units.kelvin,Td[0][0]) #parc temp profile via metpy
    CAPE_CIN = mpcalc.cape_cin(P[0].values*units.Pa,T[0].values*units.kelvin,Td[0],Parc_prof) #CAPE & CIN via metpy 
    
    #SRH @ 1km & 3 km
    u = xr.open_dataset(path2+storm16_files[0])['uinterp'].loc[:,:,y,x]
    v = xr.open_dataset(path2+storm16_files[0])['vinterp'].loc[:,:,y,x]
    z = z_data
    SRH1km = (mpcalc.storm_relative_helicity(z,u[0]*units('m/s'),v[0]*units('m/s'), depth = 1*units.km)[2]).magnitude
    SRH3km = (mpcalc.storm_relative_helicity(z,u[0]*units('m/s'),v[0]*units('m/s'), depth = 3*units.km)[2]).magnitude
        
    
    return CAPE_CIN[0].magnitude, CAPE_CIN[1].magnitude, SRH1km, SRH3km; 
      
```

```python
#CAPE, CIN, & SRH Cross Sections

#Create Storage Arrays For Necessary Variables
length = upper_bound-lower_bound
storage_CAPE = np.empty((length,length))
storage_CIN = np.empty((length,length))
storage_SRH1km = np.empty((length,length))
storage_SRH3km = np.empty((length,length))
#storage_w15m = np.empty((length,length))
#storage_w500m = np.empty((length,length))


#Computations
for x in range(length): 
    for y in tqdm(range(length), desc= "Loading..."):
        
        CAPE, CIN, SRH1km, SRH3km = computations(x,y)
        #CAPE-CIN
        storage_CAPE[y][x] = CAPE
        storage_CIN[y][x] = CIN
        #SRH1km and SRH3km
        storage_SRH1km[y][x] = SRH1km
        storage_SRH3km[y][x] = SRH3km
       

#Convert Arrays to DataArrays
storagexr_CAPE = xr.DataArray(storage_CAPE)
storagexr_CIN = xr.DataArray(storage_CIN)
storagexr_SRH1km = xr.DataArray(storage_SRH1km)
storagexr_SRH3km = xr.DataArray(storage_SRH3km)

```
