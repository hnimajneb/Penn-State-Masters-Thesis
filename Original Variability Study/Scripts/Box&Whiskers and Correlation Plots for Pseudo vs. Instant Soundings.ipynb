# **The following cells are functions needed for this script. They appear in this file because I forgot to add vertical vorticity and vertical velocity in the original interpolation scheme.**

## interp_to_grid: interpolates velocity and scalar values along the full pseudosounding trajectories, back to the original CM1 vertical grid.

## master_interpolator_t: temporal interpolation of selected fields, using the original CM1 output files (for timesteps during trajectories that are not at multiples of 30s).

## master_interpolator_s: spatial interpolation of selected fields.

```python
#Takes Raw Data Along Trajectories and Interpolates down to original z_grid in model domain
def interp_to_grid(z_points, pressure, potential_temperature, qv, u_velocity, v_velocity, w_velocity, zvort):
    
    # -------------------------------------------------------------------------------------------------------------------
   
    # INPUT VARIABLES
    
    # z_points = z_values for raw interpolated data
    # z_levels = z levels in model domain 
    
    # Raw Interpolated Data
    # pressure (Pa)
    # potential_temperature (K)
    # qv - water vapor mixing ratio (kg/kg)
    # u_velocity - u component of velocity (m/s)
    # v_velocity - v component of velocity (m/s)
    
    # -------------------------------------------------------------------------------------------------------------------
    z_levels = z_data.values
    z_points = z_points
    
    # Create Interpolator For Each Variable (interp1d creates an interpolation function 
    # using the vertical points of the raw trajectories and the associated variables - same dimensionality)
    interpolator_pres = interp1d(z_points, pressure, fill_value = 'extrapolate')
    interpolator_pot_temp = interp1d(z_points, potential_temperature, fill_value = 'extrapolate')
    interpolator_qv = interp1d(z_points, qv, fill_value = 'extrapolate')
    interpolator_u = interp1d(z_points, u_velocity, fill_value = 'extrapolate')
    interpolator_v = interp1d(z_points, v_velocity, fill_value = 'extrapolate')
    interpolator_w = interp1d(z_points, w_velocity, fill_value = 'extrapolate')
    interpolator_zvort = interp1d(z_points, zvort, fill_value = 'extrapolate')
    
    # Interpolate to Original CM1 Model Domain height levels
    interp_pres = interpolator_pres(z_levels)
    interp_pot_temp = interpolator_pot_temp(z_levels)
    interp_qv = interpolator_qv(z_levels)
    interp_u = interpolator_u(z_levels)
    interp_v = interpolator_v(z_levels)
    interp_w = interpolator_w(z_levels)
    interp_zvort = interpolator_zvort(z_levels)
    
    #return values
    return interp_pres, interp_pot_temp, interp_qv, interp_u, interp_v, interp_w, interp_zvort;
    
    
```

```python
def master_interpolator_t(analtime, time, data):
    #first we need to interpolate all fields based on the time we are currently at
    
    #Find times that are before and after the interpolation time
    before_indices = np.where(time <= analtime)[0]
    after_indices = np.where(time >= analtime)[0]
    
   
    #raise an error if the time at which interpolation occurs lies outside of the time interval of data
    if len(before_indices) == 0 or len(after_indices) == 0:
        raise ValueError('analysis time lies outside of the time interval')
    
    #indices closest before & after the interpolation time
    Bindex = before_indices[-1]
    Aindex = after_indices[0]
    
    #if time we are interpolating is the same as one of the data timesteps, just return original data at that timestep
    if time[Bindex] == analtime:
        interp_data = data[Bindex].isel(time=0)
        return interp_data
    elif time[Aindex] == analtime:
        interp_data = data[Aindex].isel(time=0)
        return interp_data
    
    #saving data locally cuts down computation time
    before_data = data[Bindex][['zvort', 'winterp']].isel(time=0)
    after_data = data[Aindex][['zvort', 'winterp']].isel(time=0)
    
    #temporal interpolation of each field individually
    
    # EXAMPLE IMPLEMENTATION: Any variable of your choosing would look like this
    #interp_u = before_data['uinterp'].values +  (
        #(after_data['uinterp'].values - 
        # before_data['uinterp'].values) *
        #(analtime - time[Bindex]) /
        #(time[Aindex] - time[Bindex])
    #)
    
    interp_zvort = before_data['zvort'].values + (
        (after_data['zvort'].values -
         before_data['zvort'].values) *
        (analtime - time[Bindex]) /
        (time[Aindex] - time[Bindex]) 
    )
    
    interp_w = before_data['winterp'].values + (
        (after_data['winterp'].values -
         before_data['winterp'].values) *
        (analtime - time[Bindex]) /
        (time[Aindex] - time[Bindex]) 
    )
    
    
    dims = ('nk', 'nj', 'ni')
    
    #returns dataset of velocity fields in a small cloud based on the previous interpolation point to save computing resources/time
    return xr.Dataset({'zvort': (dims,interp_zvort), 'winterp': (dims, interp_w)}, coords={'ni': ni,'nj': nj,'nk': nk});
    
```

```python
def master_interpolator_s(points, data):
    
    #establish the grid
    x = x_data.values
    y = y_data.values 
    z = z_data.values
    
    zvort, w = [data['zvort'].values, data['winterp'].values]
    
    #Create Interpolation Functions 
    interpolator_zvort = RegularGridInterpolator((z,y,x), zvort)
    interpolator_w = RegularGridInterpolator((z,y,x), w)
    
    #Interpolate The Data
    interp_zvort = interpolator_zvort(points)
    interp_w = interpolator_w(points)
    
    
    return interp_zvort, interp_w;
```

## Actual code doing the raw interpolation and adding it onto the pseudosounding dataset

```python
#Interpolate the Remaining Fields to the raw pseudosounding trajectories

#Storage
raw_interp_zvort = np.zeros((breakpoint, horiz_length**2))
raw_interp_w = np.zeros((breakpoint, horiz_length**2))

for i in tqdm(np.arange(0, breakpoint, 1)):
    #interpolate the data to current time
    interp_data_t = master_interpolator_t(int_time[i], time, orig_data)
    #interpolate to the points found in previous trajectory calculations (raw trajectory points)
    raw_interp_zvort[i], raw_interp_w[i] = master_interpolator_s(raw_points[i], interp_data_t)
```

```python
# Compute BWD and STP

# Bulk Wind Difference

# Height indices for ~ 0, 6 km
z_6km = (np.abs(z_data.values -6)).argmin()
z_0km = 0
# Pseudo Data
u_sfc = u[z_0km]
v_sfc = v[z_0km]
u_6km = u[z_6km]
v_6km = v[z_6km]
BWD_u = u_6km - u_sfc
BWD_v = v_6km - v_sfc
BWD = (BWD_u**2 + BWD_v**2)**(1/2)

# Inst Data
inst_u_sfc = inst_u[z_0km]
inst_v_sfc = inst_v[z_0km]
inst_u_6km = inst_u[z_6km]
inst_v_6km = inst_v[z_6km]
inst_BWD_u = inst_u_6km - inst_u_sfc
inst_BWD_v = inst_v_6km - inst_v_sfc
inst_BWD = (inst_BWD_u**2 + inst_BWD_v**2)**(1/2)

# LCLs 
# Pseudo Data
LCLs = np.zeros(10) # storage 
for x in tqdm(range(10)):
    # Select Random Location on grid 
    i = random.randint(0, P.shape[1])
    
    # Calculate LCL using metpy (units of pressure)
    LCL_prs = mpcalc.lcl(P[0,i]*units.Pa, T[0,i]*units.degK,
                         Td[0,i]*units.degK)[0]
    # Calculate LCL in height (km)
    LCL_m = mpcalc.pressure_to_height_std(LCL_prs).m
    
    # Save value 
    LCLs[x] = LCL_m
# Take average (m)
SB_LCL = LCLs.mean()

# Instant Data
inst_LCLs = np.zeros(10) # storage 
for x in tqdm(range(10)):
    # Select Random Location on grid 
    i = random.randint(0, P.shape[1])
    
    # Calculate LCL using metpy (units of pressure)
    inst_LCL_prs = mpcalc.lcl(inst_P[0,i]*units.Pa, inst_T[0,i]*units.degK,
                         inst_Td[0,i]*units.degK)[0]
    # Calculate LCL in height (km)
    inst_LCL_m = mpcalc.pressure_to_height_std(LCL_prs).m
    
    # Save value 
    inst_LCLs[x] = inst_LCL_m
# Take average (m)
inst_SB_LCL = inst_LCLs.mean()

# STP
# Pseudo 
STP = (CAPE[0]/1500)*((2000-1000*SB_LCL)/1000)*(SRH1km/150)*((200-CIN[0])/150)*(BWD/20)
# Instant
inst_STP = (inst_CAPE[0]/1500)*((2000-1000*inst_SB_LCL)/1000)*(inst_SRH1km/150)*((200-inst_CIN[0])/150)*(inst_BWD/20)
```

# Bunkers & Storm Motion For Psuedo and Instantaneous Soundings (NOT USED IN ANY FINAL WORK)

```python
#Create A Function that Computes Pressure-Weighted Mean Wind

def mean_wind(pressure, u_wind, v_wind, top, bottom = None):
    #if bottom of layer isn't specified, assume that we start from lowest level
    if bottom is None:
        bottom = 0
        top = top
        
    #input pressure column 
    P = pressure
    
    #convert winds to ground relative
    offset_u = 12.2
    offset_v = 12.5
    gr_u = u_wind + offset_u
    gr_v = v_wind + offset_v
    
    #Start Relevant Counters at 0
    u_sum = 0.0
    v_sum = 0.0
    dp_sum = 0.0
    
    #Start Computations
    for x in np.arange(bottom,top,1):
        if x == bottom:
            #initialize first data point
            p_old = P[bottom]
            u_old = gr_u[bottom]
            v_old = gr_v[bottom]
        
        else:
            #calculate dp and layer means 
            dp = p_old-P[x]
            ulayermean = (gr_u[x] + u_old)/2
            vlayermean = (gr_v[x] + v_old)/2
            
            #update sums
            u_sum += dp*ulayermean
            v_sum += dp*vlayermean
            dp_sum += dp
            
            #update old values 
            p_old = P[x]
            u_old = gr_u[x]
            v_old = gr_v[x]
            
    
    #compute mean wind
    mean_u = u_sum/dp_sum
    mean_v = v_sum/dp_sum
    
    return mean_u,mean_v;
            
```

```python
#Create a Loop That Computes Bunkers Motion For Pseudo Soundings First 

#Find Height Indices of Interest 
height = z_data.values
z_6km = (np.abs(height - 6)).argmin()
z_500m = (np.abs(height - .5)).argmin()
z_5500m = (np.abs(height - 5.5)).argmin()

#Storage For Right Mover
rm = np.zeros((P.shape[1],2))


#Loop That Computes Bunkers For Each Profile
for x in tqdm(np.arange(0,P.shape[1],1)):
    #data needed
    pres = P[:,x]
    u_wind = u[:,x]
    v_wind = v[:,x]
    
    #Step 1: Compute Mean Wind Over 500m, 6km, and 5500m to 6km
    #(first element in arrays are u-component, 2nd is v-component)
    mean_6km = np.array(mean_wind(pres, u_wind, v_wind, top = z_6km))
    mean_500m = np.array(mean_wind(pres, u_wind, v_wind, top = z_500m))
    mean_5500m_2_6km = np.array(mean_wind(pres, u_wind, v_wind, top = z_6km, bottom =z_5500m))
    
    #Step 2: Compute Shear Vector, its cross product with the k-vector, and compute shear vector magnitude 
    shear = mean_5500m_2_6km - mean_500m
    
    dtype = np.dtype([('u-component', float), ('v-component', float)])
    shear_cross = np.empty((1,1), dtype = dtype)
    shear_cross['u-component'] = shear[1]
    shear_cross['v-component'] = -shear[0]
    shear_mag = np.hypot(shear[0], shear[1])
    
    #Step 3: Compute Deviant Motion in Accordance with Bunkers (2000)
    rdev_scalar = 7.5/shear_mag
    rdev = np.empty((1,1), dtype = dtype)
    rdev['u-component'] = shear_cross['u-component']*rdev_scalar
    rdev['v-component'] = shear_cross['v-component']*rdev_scalar
    
    #Compute Storm Motion For Right Mover
    right_mover =  np.empty((1,1), dtype = dtype)
    right_mover['u-component'] = rdev['u-component']+mean_6km[0]
    right_mover['v-component'] = rdev['v-component']+mean_6km[1]
    
    #Save (1st component is u, 2nd component is v)
    rm[x] = [right_mover['u-component'], right_mover['v-component']]



        
```
